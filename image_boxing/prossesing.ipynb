```python
import cv2
import numpy as np
import os
import glob
import random

source_imgs_path = ".\\source\\source_images\\*.jpg"
source_labels_path = ".\\source\\source_labels\\*.txt"

destination_path = ".\\dst\\images\\s"

src_img_names = glob.glob(source_imgs_path)
src_lbl_names = glob.glob(source_labels_path)

i = -1
j = -1

min_ = min(len(src_img_names), len(src_lbl_names))

color_rgb_s = [(255, 0, 0)]

for _ in range(min_):
    i +=1
    j +=1
    
    img_path = src_img_names[i]
    label_path = src_lbl_names[j]



    while (".\\source\\source_labels\\" + os.path.basename(img_path)[:-4:] + ".txt") not in src_lbl_names:
        i += 1
        img_path = src_img_names[i]

    while (".\\source\\source_images\\" + os.path.basename(label_path)[:-4:] + ".jpg") not in src_img_names:
        j += 1
        label_path = src_lbl_names[j]


    if i >= len(src_img_names) or j >= len(src_lbl_names):
        break
    

    img = cv2.imread(img_path)
    img_changed = img.copy() 

    labels = np.loadtxt(label_path, dtype=float).reshape(-1, 5)
    

    labels[:,1::2] *= img.shape[0]
    labels[:,2::2] *= img.shape[1]
    
    
    x1 = (labels[:, 1] - labels[:, 3] / 2).astype(int)
    y1 = (labels[:, 2] - labels[:, 4] / 2).astype(int)
    x2 = (labels[:, 1] + labels[:, 3] / 2).astype(int)
    y2 = (labels[:, 2] + labels[:, 4] / 2).astype(int)
    
    
    for f in range(len(labels)):
        
        col_id = labels[f, 0].astype(int) 

        if col_id > len(color_rgb_s) - 1:

            for _ in range(col_id - len(color_rgb_s) + 1):
                r = random.randint(0, 255)
                g = random.randint(0, 255)
                b = random.randint(0, 255)
                color_rgb_s.append((r, g, b))

        
        color = color_rgb_s[col_id]

        img_changed = cv2.rectangle(img_changed,(x1[f], y1[f]), (x2[f], y2[f]), color, 3)
    
    new_img_path = destination_path + os.path.basename(img_path)
    
    cv2.imwrite(new_img_path, img_changed)
    
```
